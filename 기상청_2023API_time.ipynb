# 기상청_지상 시간

```python
import requests

url = 'http://apis.data.go.kr/1360000/AsosHourlyInfoService/getWthrDataList'
para ={'serviceKey' : 'MxuJY6mhvo/9EUEkXWR0mmindUYDc6YeFLi2OINqng0FsdDZsF7H8AdPzRLTn0pkGHuBkCDPS5W9FTRpYPwuRg==', 'pageNo' : '1', 'numOfRows' : '990', 'dataType' : 'JSON', 'dataCd' : 'ASOS', 'dateCd' : 'HR', 'startDt' : '20100101', 'startHh' : '01', 'endDt' : '20100601', 'endHh' : '01', 'stnIds' : '108' }

res = requests.get(url, params=para)
print(res.content)
```

```python
import json
json_file = json.loads(res.text)
```

```python
import pandas as pd
from pandas import json_normalize

df = json_normalize(json_file['response']['body']['items']['item'])
df
```

# 시각화

```python

#한글 1
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm  # 폰트 관련 용도
# 한글 폰트 세팅

plt.rc('font', family = 'NanumGothic')
mpl.rcParams['axes.unicode_minus'] = False

df['tm'] = pd.to_datetime(df['tm'])
df['ts'] = pd.to_numeric(df['ts'], downcast = 'float')
fig = plt.figure(figsize=(14,6))

ax = plt.subplot(1,1,1)
ax.plot(df['tm'], df['ts'], color = 'r', lw = 1.5, alpha = 0.7)

ax.set_ylabel('Surface Temperature', fontweight = 'bold', fontsize = 12)
ax.set_xlabel('Date 일자', fontweight = 'bold', fontsize = 12)
ax.set_title('OPEN API ASOS 온도 데이터', fontweight = 'bold', fontsize = 12) #머릿말
```
